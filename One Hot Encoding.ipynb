# One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(sparse=False)
```

```python
onehot_df = onehot.fit_transform(df[['Sex','Country','PRODUCTLINE']])
```

```python
onehot_df
```

```python
dummy_df.keys()
```

```python
 #onehot encoding gives data in numpy array. we need to converet array to data frame
onehot_df_new = pd.DataFrame(onehot_df, columns=['PRODUCTLINE_Apple',
       'PRODUCTLINE_Banana', 'PRODUCTLINE_Mango', 'Country_Belgium',
       'Sex_Female', 'Sex_Male'])    
```

```python
onehot_df_new
```

```python
onehot_df_new_merged = pd.concat([df,onehot_df_new], axis='columns')
```

```python
onehot_df_new_merged
```

```python
#drop original dataframe
onehot_df_new_final = onehot_df_new_merged.drop(['PRODUCTLINE','Country','Sex'], axis='columns')
```
