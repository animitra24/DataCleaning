# Outliers_Inter Quantile Range (IQR)

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\User\Desktop\Anindita Mitra\Dataset\Car_Sales_EDA.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
Q1 = df.quantile(0.25)
Q1
```

```python
Q3 = df.quantile(0.75)
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
lower_limit
```

```python
upper_limit = Q3 + 1.5*IQR
upper_limit
```

```python
#Identify Outliers value in a data
df[(df<lower_limit)|(df>upper_limit)]
```

```python
#Value without Outliers
df[(df>lower_limit)|(df<upper_limit)]
```
